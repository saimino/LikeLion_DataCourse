```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,14,100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4
```

```python
plt.plot(x,y1, x, y2, x, y3)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)*4")
plt.legend()
plt.title("sin plot")
```

```python
y1 = np.cos(x)
y2 = np.cos(x) * 2
y3 = np.cos(x) * 4
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)*4")
plt.legend()
plt.title("sin plot")
```

```python
y1 = np.tan(x)
y2 = np.tan(x) * 2
y3 = np.tan(x) * 4
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)*4")
plt.legend()
plt.title("sin plot")
```

```python
plt.plot(x, y1, 'o', label="sin(x)")
plt.plot(x, y2,'o', label="sin(x)*2")
plt.plot(x, y3,'o', label="sin(x)*4")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y = 'total_bill', data =tips)
plt.show() 
```

* 식사 금액은 토요일과 일요일이 많구나
* 현재데이터는 목 , 금 ,토 ,일만의 데이터가 있구나
* 토요일과 일요일을 비교해본다면 상대적으로 지출 식사금액이 많다. 

```python
# 요일별 팁, 요일별 사이즈 에 대해서.
sns.boxplot(x='day', y = 'tip', data =tips)
plt.show()
sns.boxplot(x='day', y = 'size', data =tips)
plt.show()
```

* tips

  팁은 일요일을 제외하고 평이한 수준을 보임.

  토요일은 이상값이 높은 수준에서만 많이 보임.

* size

  주말은 2명이상의 인원들이 많았음 특히, 일요일은 1명인원은 전혀 관측되지 않음.
      식사인원이 많기 때문에 식사 금액이 상대적으로 높을 수 밖에 없다.

  목요일과 금요일은 이상값이 많다.






```python
#담배피는 사람과 안이는 사람과 팁의 관계
sns.boxplot(x='day', y = 'tip',hue = 'smoker', data =tips)
plt.show()
# 담배 피는 사람이 더 많이 주는 경향이 있지만, 인과성을 찾을정도로 차이가 크지않다.
# 주말에 팁의 편차가 크며 이상값이 많다.
```

```python
#시간과 팁의 관계
sns.boxplot(x='day', y = 'tip',hue = 'time', data =tips)
plt.show()
```

```python
sns.boxplot(x='day', y = 'tip',hue = 'time', data =tips)
```
