```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,14,100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4
```

```python
plt.plot(x,y1, x, y2, x, y3)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)*4")
plt.legend()
plt.title("sin plot")
```

```python
y1 = np.cos(x)
y2 = np.cos(x) * 2
y3 = np.cos(x) * 4
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)*4")
plt.legend()
plt.title("sin plot")
```

```python
y1 = np.tan(x)
y2 = np.tan(x) * 2
y3 = np.tan(x) * 4
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)*4")
plt.legend()
plt.title("sin plot")
```

```python
plt.plot(x, y1, 'o', label="sin(x)")
plt.plot(x, y2,'o', label="sin(x)*2")
plt.plot(x, y3,'o', label="sin(x)*4")
```
